```python
from keras.models import load_model
import cv2 as cv
import numpy as np
import time
```

```python
# from keras.utils import img_to_array
```

```python
face_path = 'HaarCascadeModel/haarcascade_frontalface_default.xml'
emotion_path = 'MODEL/forEpochs_20.h5'
```

```python
face_classifier = cv.CascadeClassifier(face_path)
```

```python
emotion_classifier = load_model(emotion_path)
```

```python
emotion_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral' ,'Sad' ,'Surprise']
```

```python
capture = cv.VideoCapture(0)

while True:
    isTrue, frame = capture.read()
#     time.sleep(0.1)
    gray = cv.cvtColor(frame, cv.COLOR_RGB2GRAY)
    faces = face_classifier.detectMultiScale(gray, scaleFactor=1.3, minNeighbors = 5)
    for (x,y,w,h) in faces:
        cv.rectangle(frame,(x,y),(x+w,y+h),(255,248,240),1)
        face_detect = gray[y:y+h,x:x+w]
        face_detect = cv.resize(face_detect,(48,48),interpolation=cv.INTER_AREA)

        if np.sum([face_detect])!=0:
            fd = face_detect.astype('float')/255.0
            fd = np.array(fd)
            fd = np.expand_dims(fd,axis=0)
            prediction = emotion_classifier.predict(fd)[0]
            label=emotion_labels[prediction.argmax()]
            label_position = (x,y)
            cv.putText(frame,label,label_position,cv.FONT_HERSHEY_SIMPLEX,1,(255,255,0),2)
        else:
            cv.putText(frame,'No Faces',(30,80),cv.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2)
    cv.imshow('Emotion Detection', frame)
    if cv.waitKey(10) & 0xff == ord('q'):
        break
capture.release()
cv.destroyAllWindows()
```
